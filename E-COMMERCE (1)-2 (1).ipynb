## Standardize column names (clean & consistent)

```python
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(' ', '_')
              .str.replace('-', '_')
              .str.replace(r'[^0-9a-zA-Z_]', '', regex=True)
)
print(df.columns.tolist())
```

## Quick value-check of key columns

```python
df['payment_method'] = df['payment_method'].str.title()
```

```python
# peek at unique values for categorical-like columns
for c in ['event_type','category','payment_method','shipping_type','device_type','traffic_source','country','status']:
    if c in df.columns:
        print(f"\n== {c} ({df[c].nunique()} unique) ==")
        print(df[c].value_counts(dropna=False).head(10))
```

## Missing values diagnostics

```python
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.concat([missing, missing_pct], axis=1)
missing_df.columns = ['missing_count', 'missing_pct']
display(missing_df[missing_df['missing_count'] > 0])
```

## Deduplicate exact duplicate rows

```python
print("Exact duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True)
print("Shape after dropping duplicates:", df.shape)
```

## Parse event_time to datetime (format shown in sample: dd-mm-YYYY HH:MM)

```python
print("Sample event_time values:")
display(df['event_time'].head(5))
df['event_time'] = pd.to_datetime(df['event_time'], errors='coerce', infer_datetime_format=True)
print(df['event_time'].dtype)
```

```python
df['date'] = df['event_time'].dt.date
df['hour'] = df['event_time'].dt.hour
df['day_of_week'] = df['event_time'].dt.day_name()
df['month'] = df['event_time'].dt.month_name()
```

## Convert numeric columns to numeric dtype

```python
# list numeric-like columns based on your sample
num_cols = ['price','discount_percent','discounted_price','quantity','total_sale','profit','delivery_days','rating']

# convert safely to numeric
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# Check numeric summaries
display(df[num_cols].describe().T)
```

```python
cat_cols = ['event_type','category','payment_method','shipping_type','review','device_type','traffic_source','country','status']

for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].astype('str').str.strip().str.lower().replace({'nan': np.nan})
        
# Show cleaned uniques
for c in cat_cols:
    if c in df.columns:
        print(f"\n== {c} uniques ==")
        print(df[c].value_counts(dropna=False).head(15))
```
